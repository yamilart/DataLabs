## Standardize header names

```python
mca.columns.to_list()
```

```python
def standardize(col):
    return col.lower().replace(" ","_")
```

```python
newcolumns=[]
for col in mca.columns.to_list():
    newcolumns.append(standardize(col))
```

## Dealing with NaN values

```python
mca['vehicle_type'].unique()
```

```python
cleanmca = mca.drop(['vehicle_type'],axis=1)
```
